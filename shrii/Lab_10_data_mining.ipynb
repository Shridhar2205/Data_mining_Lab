## **Classification using a Naive Bayes Classifier**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/diabetes.csv")
```

```python
df
```

```python
y = df['Outcome']
y
```

```python
x = df.drop(columns = ['Outcome'])
def z_score(df):
    # copy the dataframe
    df_std = df.copy()
    # apply the z-score method
    for column in df_std.select_dtypes(include=np.number).columns:
        df_std[column] = (df_std[column] - df_std[column].mean()) / df_std[column].std()

    return df_std

x = z_score(x)
x
```

```python
# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=37)

# training the model on training set
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# making predictions on the testing set
y_pred = gnb.predict(X_test)

# comparing actual response values (y_test) with predicted response values (y_pred)
from sklearn import metrics
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
```

## **Regression using a Regression Tree**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/Walmart_sales.csv")
df
```

```python
X = df.drop(["Date"], axis=1)
X
```

```python
y = df["Weekly_Sales"]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=37)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state=44)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, predictions)
print("Mean Squared Error:", mse)
```
